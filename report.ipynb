### Импорт библиотек

```python
import warnings

from googletrans import Translator
from transformers import AutoTokenizer
from gliclass import GLiClassModel, ZeroShotClassificationPipeline
```

```python
# Игнорировать предупреждения
warnings.filterwarnings('ignore')
```

### Загрузка модели и токенайзера

```python
model = GLiClassModel.from_pretrained("knowledgator/gliclass-instruct-large-v1.0")
```

```python
tokenizer = AutoTokenizer.from_pretrained("knowledgator/gliclass-instruct-large-v1.0")
```

```python
pipeline = ZeroShotClassificationPipeline(model, tokenizer)
```

## Проверка работоспособности модели

```python
labels = ["positive", "negative"]
```

```python
prompts = [
    '',
    'Определите эмоциональную окраску этого сообщения.',
    'Determine the emotional coloring of this message',
    "You are an experienced sentiment analyst.",
    "You are an experienced sentiment analyst. Determine the tone (sentiment) of a user's message.'"
]
```

```python
def check_prompts(text: str, threshold=0.5):
    for prompt in prompts:
        results = pipeline(text, labels, prompt=prompt, threshold=threshold)[0]
        for r in results:
            print(f'prompt: {prompt}')
            print(r["label"], "=>", r["score"])
```

### Позитивные комментарии

```python
text = "Молодец, ты хорошо поработал."

check_prompts(text)
```

```python
text = "Well done, you did a good job."

check_prompts(text)
```

```python
text = "спасибо всем большое)"

check_prompts(text)
```

```python
text = "thank you all very much)"

check_prompts(text)
```

### Негативные комментарии

```python
text = "Молодец, ты все испортил."

check_prompts(text)
```

```python
text = "Well done, you ruined everything."

check_prompts(text)
```

```python
text = "скотина! что сказать"

check_prompts(text)
```

```python
text = "You're a beast! What can I say?"

check_prompts(text)
```

## Перевод текста с русского на английский

Для перевода текста возьмём библиотеку googletrans разработанной Google.

```python
async def translate_text(text_to_translate: str) -> str:
    async with Translator() as translator:
        result = await translator.translate(text_to_translate)
        return result.text
```

```python
await translate_text('Молодец, ты хорошо поработал.')
```

```python
await translate_text('спасибо всем большое)')
```

```python
await translate_text('Молодец, ты все испортил.')
```

```python
await translate_text('скотина! что сказать')
```
